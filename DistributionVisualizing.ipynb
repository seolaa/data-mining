```python
import pandas as pd
```

```python
game_tag = pd.read_csv('./data/game_tag.csv')
```

```python
game_tag_group = game_tag.groupby('tag').count()
```

```python
len(game_tag['tag'].unique())
```

```python
len(game_tag['game_id'].unique())
```

```python
game_tag_group = game_tag_group.sort_values('game_id', ascending=False)
game_tag_group = game_tag_group.rename(columns={'game_id':'game_count'})
```

```python
new = game_tag_group.iloc[:10]
new
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl

x = new.index
y = new['game_count']
plt.bar(x,y)
plt.axis(option='tight')
plt.show()
```

```python
user_vote = pd.read_csv('user_vote_00.csv')
user_vote
```

```python
vote = user_vote.groupby('user_id').count()
vote = vote.sort_values('game_id', ascending=False)
vote = vote.reset_index()
vote
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl

y = vote['game_id']
x = vote.index
plt.plot(x,y)
plt.xlabel('cumulative user count')
plt.ylabel('vote_count')
plt.yscale('log')
plt.show()
```

```python
user_playtime = pd.read_csv('user_playtime_00.csv')
```

```python
play = user_playtime.groupby('user_id').sum()
play = play.sort_values(by = 'play_time', ascending=False)
play = play.reset_index()
play
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl

x = play.index
y = play['play_time']
plt.plot(x,y)
plt.xlabel('cumulative user count')
plt.ylabel('play_time')
plt.show()
```
